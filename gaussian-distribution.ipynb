```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt, pi
```

```python
def function(x):
    return np.exp(-x ** 2) 
```

```python
def trapezium_rule(x, a, N):
    dx = 2 * a / N
    s = 2 * function(a)

    for i in range(N):
        s += 2 * function(x[i])

    return dx * s / 2
```

```python
def draw_graph_and_subarea(a, N, x_subarea, y_subarea, x_graph, y_graph):
    plt.figure(figsize=(12, 6))
    plt.plot(x_graph, y_graph, color='black', linewidth=1)

    for i in range(N - 1):
        plt.plot([x_subarea[i], x_subarea[i + 1]], [y_subarea[i], y_subarea[i + 1]], color='blue', linewidth=1)
        plt.plot([x_subarea[i], x_subarea[i]], [0, y_subarea[i]], color='blue')
        plt.plot([x_subarea[i + 1], x_subarea[i + 1]], [0, y_subarea[i + 1]], color='blue')
        plt.fill_between([x_subarea[i], x_subarea[i + 1]], [0, 0], [y_subarea[i], y_subarea[i + 1]], color='skyblue', alpha=0.5)

    plt.grid(True)
    plt.xlabel("x", fontsize=14)
    plt.ylabel(r'$y = e^{-x^2}$', fontsize=14)
    plt.xlim([-3, 3])
    plt.ylim([0, max(y_graph) * 1.05])
    plt.text(1.2, 0.85, f"a = {a}\nN = {N}\nSubarea = {trapezium_rule(x_subarea, a, N):.4f}\n√π = 1.77245")
    plt.title("The Bell Curve - Trapezium Rule", fontsize=16)
    plt.gca().set_aspect(aspect='auto', adjustable='datalim')

    plt.show()
```

```python
def main(N, a):
    x_subarea = np.linspace(-a, a, N)
    y_subarea = function(x_subarea)

    x_graph = np.linspace(-5, 5, 10000)
    y_graph = function(x_graph)

    draw_graph_and_subarea(a, N, x_subarea, y_subarea, x_graph, y_graph)
```

```python
main(10, 1.5)
```
