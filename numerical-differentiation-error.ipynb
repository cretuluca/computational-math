```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.sin(x)
```

```python
def df(x):
    return np.cos(x)
```

```python
def forward_errors(x, h):
    return (f(x + h) - f(x)) / h
```

```python
def centered_errors(x, h):
    return (f(x + h) - f(x - h)) / (2 * h)
```

```python
def main():
    h_values = np.logspace(-10, 0, 100)
    first_order_error = []
    second_order_error = []

    x0 = np.pi / 4

    for h in h_values:
        first_order_error.append(np.abs(forward_errors(x0, h) - df(x0)))
        second_order_error.append(np.abs(centered_errors(x0, h) - df(x0)))

    plt.xlabel('Values of h')
    plt.ylabel('Error')
    plt.loglog(h_values, first_order_error, label='Forward Difference Error')
    plt.loglog(h_values, second_order_error, label='Centered Difference Error')
    plt.loglog(h_values, h_values, label='y = h', linestyle='dashed', color='black')
    plt.loglog(h_values, h_values ** 2, label='y = hÂ²', linestyle='dashed', color='black')

    plt.legend(loc='lower right')
    plt.show()
```

```python
main()
```
